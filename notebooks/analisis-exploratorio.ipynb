```python
# Importar librerias necesarias
import pandas as pd
import re
```

```python
# 0. Cargar funciones predefinidas para limpiar datos tipo string
def normalize_string(s):
    if pd.isna(s) or str(s).lower() == 'nan':
        return pd.NA
    # Elimina espacios al inicio/final y múltiples espacios
    s = str(s).strip()
    s = re.sub(r'\s+', ' ', s)
    # Elimina espacios antes y después de comas
    s = re.sub(r'\s*,\s*', ', ', s)
    # Si queda vacío, lo convierte a NA
    if s == '':
        return pd.NA
    return s
```

```python
# 1. Cargar el dataset
df = pd.read_csv('../data/netflix_titles.csv')

# 1.1. Mostrar las primeras filas del dataset
df
```

```python
# 2. Análisis exploratorio básico
print("Dimensiones del dataset:", df.shape)
print("\nColumnas:", df.columns.tolist())
print("\nTipos de datos:\n", df.dtypes)
```

```python
# 3. Revisión de valores nulos
print("\nValores nulos por columna:\n", df.isnull().sum())

# 4. Revisión de duplicados
print("\nDuplicados por show_id:", df['show_id'].duplicated().sum())
```

```python
# 5. Ejemplo de valores únicos en columnas compuestas
print("\nEjemplo de 'cast':", df['cast'].dropna().iloc[0])
print("\nEjemplo de 'listed_in':", df['listed_in'].dropna())
print("\nEjemplo de 'duration':", df['duration'].dropna().unique()[:5])
```

```python
# 6. Limpieza de duplicados en este caso no es necesario, pero se debe hacer por buenas prácticas de programación
df = df.drop_duplicates(subset=['show_id'])

# 7. Estandarización de fechas
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# 7. Separación de 'duration' en cantidad y tipo
df[['duration_int', 'duration_type']] = df['duration'].str.extract(r'(\d+)\s*(\w+)')
df['duration_int'] = pd.to_numeric(df['duration_int'], errors='coerce')

# 8. Limpieza de espacios y nulos en texto
columnas_string = ['show_id','type','title', 'director', 'cast', 'country', 'rating', 'listed_in', 'description']
for col in columnas_string:
    df[col] = df[col].apply(normalize_string)

# 9. Mostrar resumen final
print("\nResumen después de limpieza:")
print(df.info())
df.head()

# 10. Guardar el dataset limpio
df.to_csv('../data/netflix_titles_clean.csv', index=False)
```
